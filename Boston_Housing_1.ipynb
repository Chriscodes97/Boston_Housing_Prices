```python
# Import required libraries and create dataframe
import pandas as pd
import numpy as np
df = pd.read_csv('/content/boston_housing.csv')
df.head()
```

```python
# Histogram for number of houses vs median value
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 5))
plt.hist(df['MEDV'], color = 'Tab:green')
plt.title('Boston Housing Prices Histogram', fontsize = 15)
plt.xlabel('Median value of owner-occupied homes in $1000\'s', fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.ylabel('Number of Houses', fontsize = 15)
plt.show()
```

```python
# Scatter plot for comparing target variable and crime rates
plt.figure(figsize = (8, 5))
plt.scatter(x = df['CRIM'], y = df['MEDV'], color = 'Tab:blue')
plt.xlabel('Crime Rate', fontsize = 15)
plt.ylabel('House Price', fontsize = 15)
plt.show()
```

```python
# Scatter plot for comparing target variable and lower status of population
plt.figure(figsize = (8, 5))
plt.scatter(x = df['LSTAT'], y = df['MEDV'], color = 'Tab:blue')
plt.xlabel('Percentage of Lower Status of the Population', fontsize = 15)
plt.ylabel('House Price', fontsize = 15)
plt.show()
```

```python
# Scatter plot for comparing target variable and average number of rooms per dwelling
plt.figure(figsize = (8, 5))
plt.scatter(x = df['RM'], y = df['MEDV'], color = 'Tab:blue')
plt.xlabel('Average Number of Rooms per Dwelling', fontsize = 15)
plt.ylabel('House Price', fontsize = 15)
plt.show()
```

```python
features_names = df.columns.to_list()[:-1]
X = df[features_names] # features
Y = df['MEDV'] # Target Variable
```

```python
from sklearn.model_selection import train_test_split # import for splitting the data
from sklearn.linear_model import LinearRegression #import for linear regression
```

```python
# Prepare test and train data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
# Create Linear Regression Model
my_linear_regression = LinearRegression()
my_linear_regression.fit(X_train, Y_train)
```

```python
# Test and Evaluate the Model
pred = my_linear_regression.predict(X_test)
pred
```

```python
# Create Scatterplot for the linear regression model
plt.figure(figsize = (8, 5))
plt.scatter(x = Y_test, y = pred, color = 'Tab:green')
plt.xlabel('Actual House Price', fontsize = 15)
plt.ylabel('Prediction', fontsize = 15)
plt.show()
```

```python
# Output R2 score
print('The model r2 score is: {}'.format(my_linear_regression.score(X_train, Y_train)))
```

```python
# Defining coefficients for each of the variables
print('The intercept (b_0) is  {}'.format(my_linear_regression.intercept_))
for i in range(len(features_names)):
  print('The coef (b_{}) is  {}'.format(i+1, my_linear_regression.coef_[i]))
```

```python
def MAPE(Y, pred): # MAPE: Mean Abs Percentage Error
  l1 = list(Y)
  l2 = list(pred)
  er = []
  for i in range(len(l1)):
    e = np.abs(l1[i]-l2[i])
    er.append(e/l1[i])
  return np.mean(er)
```

```python
# Import metrics needed to calculate errors
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, pred))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, pred)))
print('MAPE:', MAPE(Y_test, pred)*100, "%")
```

The regression model is able to predict approximately 76% of the Median Value of owner-occupied homes. The model contains a Mean Absolute Percentage Error of approximately 17.5%. It is a relatively good model that could be improved. Out of the variables tested, the average number of rooms per dwelling seems to have a strong relationship with house price in the scatterplot. The scatterplots also show that the house price decreases as the lower status population increases. The histogram shows that the majority of the homes lie around a median value of roughly 10-30.
